```python
import pandas as pd

# Reading the CSV file from the data_raw directory 
student = pd.read_csv("../data_raw/raw_data_sp.csv")

# Renaming columns with meaningful names
student = student.rename(columns={
    'gender': 'gender',
    'race/ethnicity': 'race_ethnicity',
    'parental level of education': 'parental_education',
    'lunch': 'Lunch',
    'math score': 'math_score',
    'reading score': 'reading_score',
    'writing score': 'writing_score',
    'test preparation course': 'test_preparation_course'
})

# Creating a DataFrame and the student performance dataset is loaded into the data frame
df_student = pd.DataFrame(student)

# Understanding the data and dimensions
print(df_student.columns)
print(df_student.shape)

# Getting the top rows
print(df_student.head())

# Getting the bottom rows
print(df_student.tail())

# Getting the Data Structure
print(df_student.dtypes)

# Converting columns to factor objects with levels
categorical_columns = ['gender', 'race_ethnicity', 'parental_education', 'Lunch', 'test_preparation_course']
for col in categorical_columns:
    df_student[col] = df_student[col].astype('category')

# Removing the unnecessary column from the data
df_student.drop(columns=['Lunch'], inplace=True)

# Summarizing the DataFrame
print(df_student.describe())

# Finding out the missing values if any
print(df_student.isna().any())

# Save the clean Data
df_student.to_csv("../data_clean/data_clean_student.csv", index=False)
```
